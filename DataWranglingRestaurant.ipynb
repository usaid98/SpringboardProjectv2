```python
import pandas as pd

file_path="C:\\Users\\Muhammad\\Downloads\\world_top_restaurants_dataset.csv"
df = pd.read_csv(file_path)
```

```python
#make a copy of the DF, snakecase, lowercase and no extra spaces
clean_df=df.copy()
clean_df.columns = (clean_df.columns.str.strip().str.lower().str.replace(" ", "_"))
print(clean_df.columns.tolist())
```

```python
#remove any whitespaces from columns
str_cols = clean_df.select_dtypes(include=["object"]).columns
for col in str_cols:
    clean_df[col] = clean_df[col].str.strip()
```

```python
#gets rid of exact duplicate rows
before_dupes = clean_df.shape[0]
clean_df = clean_df.drop_duplicates()
after_dupes = clean_df.shape[0]
```

```python
#ensures numeric columns are numeric (bad values go to NaN)
numeric_cols = [
    "michelin_stars",
    "average_price_usd",
    "established_year",
    "seating_capacity",
    "rating",
    "years_operating",
]

for col in numeric_cols:
    clean_df[col] = pd.to_numeric(clean_df[col], errors="coerce")
```

```python
#sort restaurants by rating then name
clean_df = clean_df.sort_values(
    by=["rating", "restaurant_name"],
    ascending=[False, True]
).reset_index(drop=True)
```

```python
#shows cleaned data and verify its whats we want
clean_df.info()
clean_df.head()
```

```python
#saves cleaned data into a new csv
clean_file_path = "C:\\Users\\Muhammad\\Downloads\\world_top_restaurants_dataset_cleaned.csv"
clean_df.to_csv(clean_file_path, index=False)
```
