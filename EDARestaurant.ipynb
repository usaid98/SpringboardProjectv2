```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
sns.set_context('notebook')
```

```python
df = pd.read_csv("C:\\Users\\Muhammad\\SpringboardProjects\\SpringboardProjectv2\\world_top_restaurants_dataset_cleaned.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['rating'], bins=20)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Restaurant Scores')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['average_price_usd'], bins=25)
plt.xlabel('Average Price (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Average Restaurant Price')
plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
df['country'].value_counts().head(10).plot(kind='bar')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Countries by Restaurant Count')
plt.show()
```

```python
df['price_category'].value_counts()
```

```python
plt.figure(figsize=(6,4))
df['price_category'].value_counts().plot(kind='bar')
plt.ylabel('Count')
plt.title('Restaurant Count by Price Category')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='average_price_usd', y='rating')
plt.title('Average Price vs Rating')
plt.show()
```

```python
df.groupby('average_price_usd')['rating'].mean()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='average_price_usd', y='rating')
plt.title('Rating Distribution by Average Price Range (USD)')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
numeric_df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['average_price_usd'])
plt.title('Outliers in Average Price')
plt.show()
```

```python
df['log_average_price_usd'] = np.log1p(df['average_price_usd'])
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['log_average_price_usd'], bins=25)
plt.title('Log-Transformed Average Price (USD)')
plt.show()
```
